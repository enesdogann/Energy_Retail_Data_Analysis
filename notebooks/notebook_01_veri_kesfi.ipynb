## 1. Veri Setinin Yüklenmesi

Bu bölümde gerekli Python kütüphaneleri içe aktarılmakta, Excel dosyası okunmakta, sayfa isimleri kontrol edilmekte ve her sayfa ayrı bir DataFrame olarak yüklenmektedir.


```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 20)
```

```python
# Excel dosyasının yolu
file_path = "../data/elektrik_veri.xlsx"

# Excel dosyasını açma
xls = pd.ExcelFile(file_path)

# Sayfa isimlerini görüntüleme
print("Sayfa isimleri:")
print(xls.sheet_names)
```

```python
# Her sayfayı ayrı bir DataFrame olarak yükleme

df_tahsilat = pd.read_excel(xls, sheet_name="Tahsilat")

df_tahsilat_1 = pd.read_excel(xls, sheet_name="Tahsilat 1")

df_tahakkuk = pd.read_excel(xls, sheet_name="Tahakkuk")

df_tahakkuk_1 = pd.read_excel(xls, sheet_name="Tahakkuk 1")

df_tahakkuk_2 = pd.read_excel(xls, sheet_name="Tahakkuk 2")
```

## 2. Veri Setlerinin Boyutlarının İncelenmesi

Bu bölümde her veri setinin satır ve sütun sayıları kontrol edilmektedir.

Bu kontrol sayesinde veri setlerinin eksiksiz yüklendiği ve proje dokümanındaki referans değerlerle uyumlu olduğu doğrulanmaktadır.


```python
print("Tahsilat      :", df_tahsilat.shape)
print("Tahsilat 1    :", df_tahsilat_1.shape)
print("Tahakkuk      :", df_tahakkuk.shape)
print("Tahakkuk 1    :", df_tahakkuk_1.shape)
print("Tahakkuk 2    :", df_tahakkuk_2.shape)
```

---
## 3. Veri Yapısının İncelenmesi

Bu bölümde veri setlerinin yapısı incelenmektedir.

Her veri seti için;

- sütun isimleri,
- veri tipleri,
- eksik değer durumu,
- ilk beş kayıt

incelenerek veri yapısı doğrulanmaktadır.

```python
# İlk 5 satırı görüntüleme

for isim, df in dataframes.items():
    print("=" * 80)
    print(f"{isim} Veri Seti - İlk 5 Satır")
    print("=" * 80)
    display(df.head())
```

## 4. Tanımlayıcı İstatistikler

Bu bölümde veri setlerinin temel istatistiksel özellikleri incelenmektedir.

Ortalama, minimum, maksimum, medyan, standart sapma ve çeyrek değerleri değerlendirilerek veri yapısı hakkında genel bir fikir edinilmektedir.

```python
# Tahsilat veri seti

df_tahsilat.describe()
```

```python
# Diğer veri setlerinin tanımlayıcı istatistikleri

print("Tahsilat 1 Veri Seti")
display(df_tahsilat_1.describe())

print("Tahakkuk - Hamamözü")
display(df_tahakkuk.describe())

print("Tahakkuk 1 - Gümüşhacıköy")
display(df_tahakkuk_1.describe())

print("Tahakkuk 2 - Göynücek")
display(df_tahakkuk_2.describe())
```

---
## 5. İlçelere Göre Benzersiz Müşteri Sayısı

Bu bölümde her ilçedeki benzersiz müşteri sayısı, `sozlesme_hesap_no` sütunu kullanılarak hesaplanmaktadır.

Amaç, üç ilçenin müşteri büyüklüklerini karşılaştırmak ve sonuçları referans değerlerle doğrulamaktır.

```python
# Her ilçe için benzersiz müşteri sayısını hesaplama

benzersiz_musteriler = {
    "Hamamözü": df_tahakkuk["sozlesme_hesap_no"].nunique(),
    "Gümüşhacıköy": df_tahakkuk_1["sozlesme_hesap_no"].nunique(),
    "Göynücek": df_tahakkuk_2["sozlesme_hesap_no"].nunique()
}

for ilce, musteri_sayisi in benzersiz_musteriler.items():
    print(f"{ilce}: {musteri_sayisi:,}")
```

### Değerlendirme

Analiz sonucunda en fazla benzersiz müşteriye sahip ilçe **Gümüşhacıköy** (18.190), en az müşteriye sahip ilçe ise **Hamamözü** (2.981) olarak belirlenmiştir. Bu sonuçlar proje dokümanındaki referans değerlerle uyumludur ve veri setinin doğru yüklendiğini doğrulamaktadır.

---
## 6. Tahakkuk Verilerinin Birleştirilmesi

Bu bölümde Hamamözü, Gümüşhacıköy ve Göynücek ilçelerine ait tahakkuk verileri tek bir veri setinde birleştirilmektedir.

Birleştirme işlemi sonrasında toplam kayıt sayısı kontrol edilerek veri bütünlüğü doğrulanacaktır.

```python
# Üç tahakkuk veri setini birleştirme

df_tahakkuk_tum = pd.concat(
    [df_tahakkuk, df_tahakkuk_1, df_tahakkuk_2],
    ignore_index=True
)

print("Toplam kayıt sayısı:", len(df_tahakkuk_tum))
print("Veri boyutu:", df_tahakkuk_tum.shape)
```

---
## 7. Eksik Değer Analizi

Bu bölümde birleştirilen tahakkuk veri setindeki eksik değerler incelenmektedir.

Her sütun için eksik kayıt sayısı ve eksik veri oranı hesaplanarak veri kalitesi değerlendirilmektedir.

```python
# Eksik değer analizi

eksik_degerler = pd.DataFrame({
    "Eksik Değer Sayısı": df_tahakkuk_tum.isnull().sum(),
    "Eksik Değer Oranı (%)": (
        df_tahakkuk_tum.isnull().mean() * 100
    ).round(2)
})

eksik_degerler
```

### Değerlendirme

Birleştirilen tahakkuk veri setinde yapılan inceleme sonucunda hiçbir sütunda eksik değere rastlanmamıştır.

Bu durum veri setinin analiz için uygun olduğunu ve veri bütünlüğünün korunduğunu göstermektedir.

---
## 8. Negatif kWh Değerlerinin İncelenmesi

Bu bölümde `kwh` sütununda bulunan negatif tüketim değerleri tespit edilmektedir.

Negatif değerler doğrudan silinmeden önce sayıları ve örnek kayıtları incelenecektir.

```python
# Negatif kWh değerlerini filtreleme

negatif_kwh = df_tahakkuk_tum[df_tahakkuk_tum["kwh"] < 0]

print("Negatif kWh kayıt sayısı:", len(negatif_kwh))
print("Negatif kWh oranı (%):", round(len(negatif_kwh) / len(df_tahakkuk_tum) * 100, 4))

display(negatif_kwh.head())
```

### Değerlendirme

Analiz sonucunda **151 adet negatif kWh değeri** tespit edilmiştir. Bu kayıtlar toplam veri setinin yaklaşık **%0,013'ünü** oluşturmaktadır.

Negatif tüketim değerleri doğrudan hatalı veri olarak değerlendirilmemiştir. Negatif tüketim kayıtları sayaç düzeltmeleri, mahsuplaşmalar, fatura iptalleri veya veri giriş süreçleriyle ilişkili olabilir. Ancak mevcut veri seti tek başına bu kayıtların kesin nedenini açıklamamaktadır.

Bu nedenle bu aşamada herhangi bir kayıt silinmemiş, yalnızca tespit edilerek raporlanmıştır.

---
## 9. Aykırı Değer (Outlier) Analizi

Bu bölümde `kwh` sütunundaki aykırı değerler IQR (Interquartile Range) yöntemi kullanılarak tespit edilmektedir.

Aykırı değerler analiz edilmek amacıyla belirlenmiş olup bu aşamada veri setinden çıkarılmamıştır.

```python
# IQR yöntemi ile aykırı değer analizi

Q1 = df_tahakkuk_tum["kwh"].quantile(0.25)
Q3 = df_tahakkuk_tum["kwh"].quantile(0.75)

IQR = Q3 - Q1

alt_sinir = Q1 - 1.5 * IQR
ust_sinir = Q3 + 1.5 * IQR

aykiri_degerler = df_tahakkuk_tum[
    (df_tahakkuk_tum["kwh"] < alt_sinir) |
    (df_tahakkuk_tum["kwh"] > ust_sinir)
]

print(f"Alt sınır : {alt_sinir:.2f}")
print(f"Üst sınır : {ust_sinir:.2f}")
print(f"Aykırı değer sayısı : {len(aykiri_degerler)}")
print(f"Aykırı değer oranı : %{len(aykiri_degerler) / len(df_tahakkuk_tum) * 100:.2f}")

display(aykiri_degerler.head())
```

### Değerlendirme

IQR yöntemi sonucunda **48.554 adet aykırı değer** tespit edilmiştir. Bu kayıtlar toplam veri setinin yaklaşık **%4,09'unu** oluşturmaktadır.

Aykırı olarak belirlenen değerler istatistiksel yönteme göre hesaplanmıştır. Ancak enerji tüketim verilerinde yüksek tüketimli aboneler doğal olarak bu aralığın dışında kalabilmektedir.

Bu nedenle aykırı değerler veri setinden çıkarılmamış, yalnızca analiz kapsamında raporlanmıştır.

---
## 10. Hesap Sınıfına Göre Elektrik Tüketim Analizi

Bu bölümde hesap sınıflarına göre elektrik tüketim değerlerinin temel istatistikleri hesaplanmaktadır.

Her hesap sınıfı için;

- Kayıt sayısı
- Ortalama tüketim (kWh)
- Medyan tüketim (kWh)
- Standart sapma

değerleri hesaplanarak tüketim davranışları karşılaştırılmaktadır.

```python
# Hesap sınıfına göre temel istatistikler

hesap_sinifi_istatistik = (
    df_tahakkuk_tum
    .groupby("Hesap Sınıfı")["kwh"]
    .agg(
        Kayit_Sayisi="count",
        Ortalama_kWh="mean",
        Medyan_kWh="median",
        Standart_Sapma="std"
    )
    .round(2)
    .sort_values("Ortalama_kWh", ascending=False)
)

hesap_sinifi_istatistik
```

### Değerlendirme

Hesap sınıflarına göre yapılan analiz sonucunda en yüksek ortalama elektrik tüketiminin **Karayolları Genel Müdürlüğü Aydınlatma**, **Arıtma Tesisleri**, **Lisanssız Üreticiler** ve **Sanayi** abonelerinde olduğu görülmüştür.

Mesken ve benzeri bireysel abonelerde ortalama tüketim daha düşük seviyelerde gerçekleşirken, kurumsal ve endüstriyel abonelerde tüketim değerleri belirgin şekilde yüksektir. Bu sonuçlar hesap sınıfları arasındaki tüketim farklılıklarını açıkça göstermektedir.

Ayrıca IQR sınırları tüm hesap sınıfları birlikte değerlendirilerek hesaplanmıştır. Hesap sınıflarının tüketim yapıları farklı olduğundan, ileri analizlerde aykırı değerlerin hesap sınıfı bazında ayrıca incelenmesi daha sağlıklı olacaktır.
